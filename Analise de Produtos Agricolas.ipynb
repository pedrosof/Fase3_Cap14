# Análise de Produtos Agrícolas

Esta análise explora os outliers usando boxplots e a correlação entre variáveis como Nitrogênio (N), Fósforo (P), Potássio (K), Temperatura, Umidade e Precipitação.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
    
```

```python

# Substitua o caminho pelo caminho correto do seu arquivo CSV
arquivo_csv = '/Users/pedrosof/Documents/FIAP/Trabalhos/Fase3_Cap14/Atividade_Cap_14_produtos_agricolas.csv'
dados = pd.read_csv(arquivo_csv)

# Exibir as primeiras linhas do dataset
dados.head()
    
```

# Boxplots

```python
# Reformatar os dados para o formato "longo"
dados_long = pd.melt(dados, value_vars=['N', 'P', 'K', 'humidity', 'temperature', 'rainfall'],
                     var_name='Variável', value_name='Valor')

# Criar um boxplot com todas as variáveis no mesmo gráfico
plt.figure(figsize=(10, 6))
sns.boxplot(x='Variável', y='Valor', data=dados_long, hue='Variável', dodge=False)
plt.title('Boxplot de Variáveis Agrícolas')
plt.xlabel('Variáveis')
plt.ylabel('Valor')
plt.legend([], [], frameon=False)
plt.show()
```

## **Boxplot de Variáveis Agrícolas**

Este gráfico de **boxplot** mostra as distribuições de várias variáveis agrícolas, incluindo **Nitrogênio (N)**, **Fósforo (P)**, **Potássio (K)**, **Umidade (humidity)**, **Temperatura (temperature)** e **Precipitação (rainfall)**. O gráfico fornece uma visão clara sobre a dispersão, a mediana e possíveis outliers de cada variável.

## Descrição das Partes do Boxplot:
- **Caixa (box)**: Representa o intervalo interquartílico (IQR), ou seja, os valores entre o primeiro quartil (Q1 - 25%) e o terceiro quartil (Q3 - 75%). Isso cobre 50% dos dados.
- **Linha no meio da caixa**: Representa a **mediana** (valor central da distribuição).
- **"Whiskers" (linhas que se estendem a partir da caixa)**: Estendem-se até os valores mais extremos que ainda estão dentro de 1,5 vezes o intervalo interquartílico.
- **Pontos fora das "whiskers"**: Esses são os **outliers**, ou seja, valores que estão fora do intervalo típico.

## Análise das Variáveis:
1. **Nitrogênio (N)**:
   - A distribuição de **Nitrogênio (N)** parece ter uma mediana que é levemente inclinada para o lado inferior, com alguns **outliers** no topo, indicando que há alguns campos com níveis de nitrogênio significativamente mais altos que a maioria.
   - A caixa é relativamente pequena, o que indica que os níveis de nitrogênio estão mais concentrados dentro de uma faixa específica.
  
2. **Fósforo (P)**:
   - A variável **Fósforo (P)** apresenta uma distribuição um pouco mais ampla, com a mediana também tendendo para a parte inferior da caixa.
   - A dispersão é maior, e há outliers visíveis tanto na parte inferior quanto superior, sugerindo que os níveis de fósforo variam mais amplamente do que o nitrogênio.

3. **Potássio (K)**:
   - A variável **Potássio (K)** também mostra uma faixa ampla de distribuição, com alguns outliers.
   - A caixa é mais alongada, indicando que os níveis de potássio são mais dispersos dentro do conjunto de dados.

4. **Umidade (humidity)**:
   - A variável de **Umidade** parece estar bem distribuída com uma faixa moderada e menos outliers comparado ao nitrogênio e fósforo.
   - A mediana está mais próxima do centro, sugerindo que a umidade está distribuída de forma mais simétrica.

5. **Temperatura (temperature)**:
   - A **temperatura** apresenta uma dispersão menor em comparação com outras variáveis, com uma distribuição relativamente simétrica e poucos outliers.
   - Isso sugere que a temperatura foi mais consistentemente medida ou controlada entre os campos.
  
6. **Precipitação (rainfall)**:
   - A **precipitação** tem uma grande dispersão, com uma mediana inclinada para baixo e muitos **outliers** na parte superior, indicando que há variações climáticas significativas em diferentes regiões.

## Conclusão:
Esse **boxplot** é útil para visualizar a dispersão e as diferenças entre as variáveis agrícolas. As variáveis **N**, **P**, **K** tendem a ter maiores variações em relação a umidade, temperatura e precipitação, com outliers significativos indicando áreas que podem necessitar de maior atenção ou investigação.

# Histograma KDE

```python
# Variáveis a serem analisadas
variaveis = ['N', 'P', 'K', 'humidity', 'temperature', 'rainfall']

# Criar gráficos de histograma com KDE para cada variável
plt.figure(figsize=(12, 10))
for i, var in enumerate(variaveis, 1):
    plt.subplot(2, 3, i)  # 2x3 grid para 6 variáveis
    sns.histplot(dados[var], kde=True, color="blue")
    plt.title(f'Histograma e KDE de {var}')
plt.tight_layout()
plt.show()
```

O KDE é uma estimativa de densidade que suaviza os dados, permitindo uma visualização mais fluida da distribuição dos valores.

## Descrição das Variáveis:
1. **Nitrogênio (N)**:
   - O histograma de **Nitrogênio (N)** apresenta uma distribuição levemente assimétrica, com uma maior concentração de dados no lado inferior. A curva KDE sugere que a maioria dos campos apresenta níveis de nitrogênio em uma faixa baixa, com uma cauda longa indicando alguns valores altos.
  
2. **Fósforo (P)**:
   - A distribuição de **Fósforo (P)** mostra uma concentração maior de valores em níveis baixos, mas a curva KDE apresenta uma distribuição mais ampla do que o Nitrogênio, sugerindo uma maior variabilidade nos níveis de fósforo.
  
3. **Potássio (K)**:
   - A variável **Potássio (K)** tem uma distribuição similar ao fósforo, com uma leve assimetria, mas sem picos extremos visíveis. A curva KDE é mais suave, indicando uma distribuição mais homogênea dos valores.
  
4. **Umidade (humidity)**:
   - O gráfico de **Umidade (humidity)** mostra uma distribuição mais simétrica, com a maioria dos valores concentrados em torno do centro. A curva KDE é suave e simétrica, indicando uma variação mais uniforme da umidade entre os campos.
  
5. **Temperatura (temperature)**:
   - A distribuição de **Temperatura (temperature)** é bastante centrada, com uma curva KDE que confirma que os valores de temperatura não variam tanto, e a maioria dos dados está concentrada em torno de um intervalo central.
  
6. **Precipitação (rainfall)**:
   - A variável **Precipitação (rainfall)** apresenta uma distribuição mais dispersa, com a curva KDE mostrando uma cauda mais longa, indicando que há uma maior variabilidade na precipitação. A maioria dos campos parece estar em uma faixa baixa de precipitação, com alguns outliers indicando níveis muito altos.

## Conclusão:
Os histogramas com KDE são úteis para visualizar a distribuição dos dados de uma forma mais fluida e detalhada. As variáveis de nutrientes (N, P, K) mostram maior variabilidade entre os campos analisados, enquanto as condições ambientais como umidade e temperatura são mais uniformes, com exceção da precipitação, que apresenta maior variabilidade.


# Correlação por Mapa de Calor

```python
# Matriz de correlação entre as variáveis
correlacao = dados[['N', 'P', 'K', 'humidity', 'temperature', 'rainfall']].corr()

# Criar o heatmap para exibir a correlação entre as variáveis
plt.figure(figsize=(8, 6))
sns.heatmap(correlacao, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Mapa de Calor da Correlação entre Variáveis')
plt.show()
```

## **Mapa de Calor da Correlação entre Variáveis**

Este gráfico de **heatmap** mostra a **correlação** entre as variáveis **Nitrogênio (N)**, **Fósforo (P)**, **Potássio (K)**, **Umidade (humidity)**, **Temperatura (temperature)** e **Precipitação (rainfall)**. O coeficiente de correlação varia entre **-1 e 1**, onde:
- **1**: Correlação positiva perfeita (as variáveis aumentam ou diminuem juntas).
- **0**: Sem correlação (as variáveis não têm relação linear).
- **-1**: Correlação negativa perfeita (quando uma aumenta, a outra diminui).

## Descrição das Correlações:
1. **Correlação entre Nitrogênio (N) e Fósforo (P)**:
   - A correlação entre **Nitrogênio (N)** e **Fósforo (P)** é levemente positiva, indicando que, em geral, à medida que os níveis de nitrogênio aumentam, os níveis de fósforo também tendem a aumentar, embora essa relação não seja muito forte.

2. **Correlação entre Potássio (K) e Fósforo (P)**:
   - A correlação entre **Potássio (K)** e **Fósforo (P)** também é positiva, embora seja uma relação moderada. Isso indica que níveis mais altos de fósforo tendem a estar associados a níveis mais altos de potássio nos campos, o que pode refletir práticas de fertilização semelhantes.

3. **Correlação entre Umidade (humidity) e Precipitação (rainfall)**:
   - A correlação de **0.094** entre **Umidade** e **Precipitação** é muito fraca, sugerindo que, no conjunto de dados analisado, a precipitação não está diretamente associada a um aumento significativo da umidade. Outros fatores, como temperatura e características do solo, podem estar influenciando os níveis de umidade de forma mais pronunciada.

4. **Correlação entre Temperatura (temperature) e Umidade (humidity)**:
   - A **temperatura** e a **umidade** apresentam uma leve correlação negativa, sugerindo que temperaturas mais altas tendem a estar associadas a uma menor umidade nos campos. Essa relação é consistente com a expectativa de que temperaturas elevadas podem secar o solo.

5. **Correlação entre Nitrogênio (N) e Potássio (K)**:
   - A correlação positiva entre **Nitrogênio** e **Potássio** indica que os campos com níveis mais elevados de um desses nutrientes tendem a ter níveis mais altos do outro, embora a relação não seja muito intensa.

## Conclusão:
O **heatmap** de correlação fornece uma visão clara das relações lineares entre as variáveis do conjunto de dados. As correlações entre os nutrientes são positivas, mas moderadas, enquanto as variáveis climáticas, como umidade e temperatura, mostram padrões previsíveis. A correlação muito baixa entre **precipitação** e **umidade** indica que a relação entre essas variáveis pode ser influenciada por outros fatores que precisam ser investigados mais a fundo.


# Pairplot (Gráfico de Dispersão entre Pares)

```python
# Criar o pairplot com as variáveis de interesse
sns.pairplot(dados[['N', 'P', 'K', 'humidity', 'temperature', 'rainfall']], diag_kind='kde', markers='o')
plt.suptitle('Análise de Dispersão entre Variáveis', y=1.02)
plt.show()
```

## **Pairplot de Dispersão entre Variáveis**

O gráfico de **pairplot** é útil para visualizar a relação entre variáveis duas a duas. Ele exibe gráficos de dispersão (scatter plots) para todas as combinações de variáveis, além de gráficos de densidade na diagonal (KDE). Isso permite observar visualmente como as variáveis se correlacionam e identificar padrões como correlações, clusters e outliers.

## Descrição das Relações Entre Variáveis:

1. **Relação entre Nitrogênio (N) e Fósforo (P)**:
   - A dispersão entre **Nitrogênio (N)** e **Fósforo (P)** mostra uma leve tendência positiva, onde campos com mais nitrogênio tendem a ter mais fósforo, mas não há uma correlação forte. A relação parece difusa e sem um padrão linear claro.

2. **Relação entre Nitrogênio (N) e Potássio (K)**:
   - Há uma maior dispersão entre **Nitrogênio (N)** e **Potássio (K)**, sugerindo uma possível tendência positiva, mas com uma variação considerável. Não há um padrão linear claro, mas os pontos estão distribuídos de forma relativamente homogênea.

3. **Relação entre Fósforo (P) e Potássio (K)**:
   - A relação entre **Fósforo (P)** e **Potássio (K)** também parece seguir um padrão semelhante, com uma leve tendência positiva, mas a dispersão dos pontos sugere que as variáveis não estão fortemente correlacionadas.

4. **Relação entre Umidade (humidity) e Temperatura (temperature)**:
   - A dispersão entre **Umidade** e **Temperatura** mostra uma leve relação negativa, onde, em geral, campos com temperaturas mais altas tendem a ter níveis de umidade mais baixos. Essa relação está de acordo com as expectativas em relação às condições ambientais.

5. **Relação entre Precipitação (rainfall) e Umidade (humidity)**:
   - Há uma relação mais clara e positiva entre **Precipitação** e **Umidade**, onde campos com maior precipitação tendem a ter mais umidade. Esse padrão é consistente com o esperado, já que a precipitação diretamente afeta os níveis de umidade no solo.

6. **Relação entre Temperatura (temperature) e Precipitação (rainfall)**:
   - A relação entre **Temperatura** e **Precipitação** é menos evidente, com os dados distribuídos de maneira mais aleatória, sugerindo que essas duas variáveis podem não estar diretamente relacionadas nos dados apresentados.

## Conclusão:
O gráfico de **pairplot** oferece uma maneira rápida de visualizar como variáveis numéricas estão relacionadas entre si. Ele sugere padrões de correlação leve entre algumas variáveis, como os nutrientes, mas os padrões são mais fortes entre as variáveis ambientais, como umidade e precipitação. Este gráfico é útil para explorar visualmente os dados e levantar hipóteses sobre relações entre as variáveis.


# Gráfico de Violino para comparar as distribuições das variáveis numéricas em relação ao tipo de cultura

# Gráfico de Barras Empilhadas

```python
# Podemos agrupar as categorias em relação à precipitação
dados['rainfall_bin'] = pd.cut(dados['rainfall'], bins=4, labels=['Baixa', 'Média', 'Alta', 'Muito Alta'])

# Criar o gráfico de barras empilhadas para visualizar a distribuição de tipos de cultura em relação à precipitação
plt.figure(figsize=(12, 8))  # Aumentar o tamanho da imagem
sns.countplot(x='rainfall_bin', hue='label', data=dados)
plt.title('Distribuição de Tipos de Cultura por Intervalo de Precipitação')
plt.xlabel('Intervalo de Precipitação')
plt.ylabel('Contagem de Tipos de Cultura')

# Ajustar a posição da legenda para fora do gráfico
plt.legend(title='Tipos de Cultura', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar o layout da figura
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

## **Distribuição de Tipos de Cultura por Intervalo de Precipitação**

Este gráfico de **barras empilhadas** mostra a distribuição de diferentes tipos de cultura em relação a intervalos de precipitação classificados em quatro categorias: **Baixa**, **Média**, **Alta** e **Muito Alta**. A visualização permite observar como os diferentes tipos de cultura se distribuem em condições de precipitação variadas, o que pode fornecer insights sobre a adaptabilidade de certas culturas a diferentes níveis de chuva.

## Descrição das Distribuições:

### 1. Distribuição de Culturas em Precipitação Baixa:
   - No intervalo de **precipitação baixa**, observa-se que algumas culturas têm uma maior presença, indicando que essas culturas podem ser mais tolerantes à seca ou cultivadas em regiões com menos chuvas.
   - Certas culturas estão sub-representadas, sugerindo que podem não ser viáveis em condições de baixa precipitação.

### 2. Distribuição de Culturas em Precipitação Média:
   - No intervalo de **precipitação média**, há uma distribuição mais equilibrada entre os tipos de cultura. A maioria das culturas parece prosperar nesse intervalo, o que pode indicar que condições de precipitação moderada são mais adequadas para o cultivo de várias culturas.
   - A quantidade de cada cultura não varia muito nesse intervalo, sugerindo que a precipitação média pode ser favorável para o crescimento de muitas culturas.

### 3. Distribuição de Culturas em Precipitação Alta:
   - A categoria de **precipitação alta** mostra uma redução em algumas culturas que prosperaram em precipitações médias, enquanto outras culturas aumentam sua presença.
   - Isso pode sugerir que certas culturas têm uma maior tolerância à umidade e podem ser preferencialmente cultivadas em regiões com chuvas mais frequentes ou intensas.

### 4. Distribuição de Culturas em Precipitação Muito Alta:
   - No intervalo de **precipitação muito alta**, há uma concentração de poucas culturas dominantes, enquanto outras quase desaparecem. Isso pode indicar que apenas certas culturas são capazes de sobreviver ou prosperar em condições de chuva extrema.
   - As culturas que continuam presentes nesse intervalo são, possivelmente, aquelas mais resistentes à saturação de água no solo.

## Conclusão:
O gráfico de barras empilhadas evidencia a relação entre precipitação e adaptabilidade das culturas. Ele demonstra que diferentes culturas têm preferências distintas em relação aos níveis de chuva, com algumas culturas sendo mais adequadas para condições de baixa precipitação, enquanto outras prosperam em ambientes com chuvas intensas. Essa informação pode ser útil para práticas agrícolas em diferentes regiões, dependendo dos padrões de precipitação locais.
